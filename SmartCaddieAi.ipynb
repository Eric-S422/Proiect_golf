RAND NOU

```python
new_row_data = {}
for column in data.columns:
    if column == 'Smash factor':
        new_row_data[column] = [np.nan]  # initializez cu NaN pentru calcul mai tarziu
        continue
    if column == 'Fade/Draw':
        new_row_data[column] = [None]  # initializez cu None pentru calcul mai tarziu
        continue

    value = input(f"Enter value for '{column}': ")
    try:
        if column in ['Name', 'Club']:
            new_row_data[column] = [str(value)]
        else:
            if '.' in value:
                new_row_data[column] = [float(value)]
            else:
                new_row_data[column] = [int(value)]
    except ValueError:
        new_row_data[column] = [str(value)]

# Convertie dictionat la dataFrame
new_row_df = pd.DataFrame(new_row_data)

# adaug linia la setul de date
data = pd.concat([data, new_row_df], ignore_index=True)

last_idx = data.index[-1] # indexul coloanei noi

# calcul smash factor din program
ball_speed = data.loc[last_idx, 'Ball Speed']
club_speed = data.loc[last_idx, 'Club Speed']

#verific daca pot calcula smash factor si nu am valori nule
if pd.notna(ball_speed) and pd.notna(club_speed) and club_speed != 0:
    data.loc[last_idx, 'Smash factor'] = (ball_speed / club_speed).round(2)
else:
    data.loc[last_idx, 'Smash factor'] = np.nan #  NaN daca calculele nu sunt posibile

# Calcul 'Fade/Draw' coloana noua
face_to_path = data.loc[last_idx, 'Face to path']
club_path = data.loc[last_idx, 'Club path']
curve = data.loc[last_idx, 'Curve']

if pd.notna(face_to_path) and pd.notna(club_path) and pd.notna(curve):
    if (face_to_path > 0) & (club_path < 0) & (curve < 15) & (curve > -15):
        data.loc[last_idx, 'Fade/Draw'] = "Fade"
    elif (face_to_path < 0) & (club_path > 0) & (curve < 15) & (curve > -15):
        data.loc[last_idx, 'Fade/Draw'] = "Draw"
    elif (face_to_path > 0) & (club_path > 0) & (curve < 15) & (curve > -15):
        data.loc[last_idx, 'Fade/Draw'] = "Push"
    elif (face_to_path < 0) & (club_path < 0) & (curve < 15) & (curve > -15):
        data.loc[last_idx, 'Fade/Draw'] = "Pull"
    elif (face_to_path > 0) & (club_path < 0) & (curve > 15):
        data.loc[last_idx, 'Fade/Draw'] = "Slice"
    elif (face_to_path < 0) & (club_path > 0) & (curve < -15):
        data.loc[last_idx, 'Fade/Draw'] = "Hook"
    elif (face_to_path < 1) & (face_to_path > -1) & (club_path < 1) & (club_path > -1):
        data.loc[last_idx, 'Fade/Draw'] = "Square"
    else:
        data.loc[last_idx, 'Fade/Draw'] = None # If none of the conditions match
else:
    data.loc[last_idx, 'Fade/Draw'] = None # Assign None if calculation not possible

print("New row added successfully!")
display(data.tail())
```

```python
club = input(f"What club do you want to check the scores?")
target_carry = float(input(f"Enter the target carry distance: "))
target_total = float(input(f"Enter the target total distance: "))
target_side = float(input(f"Enter target side:"))
target_ball_speed = float(input(f"Enter target ball speed: "))
target_club_speed = float(input(f"Enter target club speed: "))
target_smash_factor = float(input(f"Enter target smash factor: "))
target_attack_angle = float(input(f"Enter target attack angle: "))
target_face_path = float(input(f"Enter target face to path: "))
target_club_path = float(input(f"Enter target club path: " ))
print("Let's check your performance: \n")
dev_carry_total = float(input(f"What deviation do you want to set for the carry & distance: "))
dev_side = float(input(f"What deviation do you want to set for the side: "))
dev_ball_speed = float(input(f"What deviation do you want to set for the ball speed: "))
dev_club_speed = float(input(f"What deviation do you want to set for the club speed: "))
dev_smash_factor = float(input(f"What deviation do you want to set for the smash factor: "))
dev_attack_angle = float(input(f"What deviation do you want to set for the attack angle: "))
dev_face_club_path = float(input(f"What deviation do you want to set for the face to path and club path: "))

score_carry = 0
score_total = 0
score_side = 0
score_ball_speed = 0
score_club_speed = 0
score_smash = 0
score_attack = 0
score_face_path = 0
score_club_path = 0

# filtrez data pentru crosa aleasa
club_data = data[data['Club'] == club]

for i in range(len(club_data)):
  if (club_data['Carry'].iloc[i] < target_carry + dev_carry_total) and (club_data['Carry'].iloc[i] > target_carry - dev_carry_total):
    score_carry += 1
  if (club_data['Total'].iloc[i] < target_total + dev_carry_total) and (club_data['Total'].iloc[i] > target_total - dev_carry_total):
    score_total += 1
  # valoare absoluta??? relatie pentru side club si face path
  if (abs(club_data['Side'].iloc[i] - target_side) < dev_side):
    score_side += 1
  if (club_data['Ball Speed'].iloc[i] < target_ball_speed + dev_ball_speed) and (club_data['Ball Speed'].iloc[i] > target_ball_speed - dev_ball_speed):
    score_ball_speed += 1
  if (club_data['Club Speed'].iloc[i] < target_club_speed + dev_club_speed) and (club_data['Club Speed'].iloc[i] > target_club_speed - dev_club_speed):
    score_club_speed += 1
  if (club_data['Smash factor'].iloc[i] < target_smash_factor + dev_smash_factor) and (club_data['Smash factor'].iloc[i] > target_smash_factor - dev_smash_factor):
    score_smash += 1
  if (club_data['Attack angle'].iloc[i] < target_attack_angle + dev_attack_angle) and (club_data['Attack angle'].iloc[i] > target_attack_angle - dev_attack_angle): # Corrected dev_smash_factor to dev_attack_angle
    score_attack += 1
  if (club_data['Face to path'].iloc[i] < target_face_path + dev_face_club_path) and (club_data['Face to path'].iloc[i] > target_face_path - dev_face_club_path):
    score_face_path += 1
  if (club_data['Club path'].iloc[i] < target_club_path + dev_face_club_path) and (club_data['Club path'].iloc[i] > target_club_path - dev_face_club_path):
    score_club_path += 1

num_shots = len(club_data) # Nr total de lovituri pe crosa aleasa

if num_shots == 0:
    print(f"No shots found for club: {club}")
else:
    score_overall = score_carry + score_total + score_side + score_ball_speed + score_club_speed + score_smash + score_attack + score_face_path + score_club_path # Corrected typo
    overall_len = 9 * num_shots # 9 metrics, multiplied by the number of shots for the selected club

    print("Your carry distance score is ", score_carry, " out of ", num_shots, " and your percentage is ", round((score_carry/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your total distance score is ", score_total, " out of ", num_shots, " and your percentage is ", round((score_total/num_shots*100),2) if num_shots > 0 else 0),"%"
    print("Your side score is ", score_side, " out of ", num_shots, " and your percentage is ", round((score_side/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your ball speed score is ", score_ball_speed, " out of ", num_shots, " and your percentage is ", round((score_ball_speed/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your club speed score is ", score_club_speed, " out of ", num_shots, " and your percentage is ", round((score_club_speed/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your smash factor score is ", score_smash, " out of ", num_shots, " and your percentage is ", round((score_smash/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your attack angle score is ", score_attack, " out of ", num_shots, " and your percentage is ", round((score_attack/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your face to path score is ", score_face_path, " out of ", num_shots, " and your percentage is ", round((score_face_path/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your club path score is ", score_club_path, " out of ", num_shots, " and your percentage is ", round((score_club_path/num_shots*100),2) if num_shots > 0 else 0,"%")
    print("Your overall score based on this criterias is: ", score_overall, " out of ", overall_len , " and your overall percentage is ", round((score_overall/overall_len*100),2) if overall_len > 0 else 0,"%")
```
